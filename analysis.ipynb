```python
# create column of network names
```

```python
name_list = list(data['label_full_name'])
name_list
```

```python
netname_list = []
for name in name_list:
    if ("TempPar" in name) or ("SomMotA" in name):
        new_name = name.split("_")[:-1]
        net_name = '_'.join(new_name[-2:])
        print(net_name)
        netname_list.append(net_name)
    else:
        new_name = name.split("_")[:-2]
        net_name = '_'.join(new_name[-2:])
        print(net_name)
        netname_list.append(net_name)
```

```python
data["network"] = netname_list
data
```

```python
sns.scatterplot(data=data, x="mean w-score", y="perc_suprathreshold", hue="network", legend=False)
```
